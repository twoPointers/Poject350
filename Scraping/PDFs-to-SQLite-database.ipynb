```python
courseNames = dict()
courseNames = {
    'CSE133' : 'Structured Programming Language',
    'CSE134' : 'Structured Programming Language Lab',
    'CSE143' : 'Discrete Mathematics',
    'CSE144' : 'Discrete Mathematics Lab',
    'EEE109D' : 'Electrical Circuits',
    'EEE110D' : 'Electrical Circuits Lab',
    'ENG101D' : 'Effective Communication in English',
    'ENG102D' : 'English Language Lab 1',
    'MAT102D' : 'Matrices, Vector Analysis and Geometry',

    'CSE137' : 'Data Structure',
    'CSE138' : 'Data Structure Lab',
    'EEE111D' : 'Electronic Devices and Circuits',
    'EEE112D' : 'Electronic Devices and Circuits Lab',
    'IPE106D' : 'Engineering Graphics',
    'IPE108D' : 'Workshop Practice',
    'PHY103D' : 'Mechanics, Wave, Heat & Thermodynamics',
    'MAT103D' : 'Calculus',
    'BNG105D' : 'Bangla Language',
    'CSE150' : 'Project Work 1',

    'CSE233' : 'Object Oriented Programming Language',
    'CSE234' : 'Object Oriented Programming Language Lab',
    'CSE237' : 'Algorithm Design & Analysis',
    'CSE238' : 'Algorithm Design & Analysis Lab',
    'BUS203D' : 'Cost and Management Accounting',
    'PHY202D' : 'Basic Physics Lab',
    'PHY207D' : 'Electromagnetism, Optics & Modern Physics',
    'STA202D' : 'Basic Statistics & Probability',

    'CSE239' : 'Numerical Analysis',
    'CSE240' : 'Numerical Analysis Lab',
    'CSE245' : 'Theory of Computation',
    'CSE249' : 'Ethics and Cyber Law',
    'CSE250' : 'Project Work 2',
    'CSE252' : 'Competitive Programming',
    'ECO105D' : 'Principles of Economics',
    'EEE201D' : 'Digital Logic Design',
    'EEE202D' : 'Digital Logic Design Lab',
    'MAT204D' : 'Complex Variables, Laplace Transform and Fourier Series'
}

def getCourse(courseString):
    Course = {
        "CourseID": "",
        "CourseName": "",
        "Semester": "",
        "Credit": 0.00,
        "isMajor": True,
        "isLab": True
    }
    c = courseString.split()
    Course["CourseID"] = c[0]
    Course["CourseName"] = courseNames.get(c[0])
    Course["Semester"] = c[2][0]
    Course["Credit"] = float(c[1])
    code = c[0][3:6]
    Course["isMajor"] = False if (int(code[1]) < 2) else True
    Course["isLab"] = True if (int(code) % 2 == 0) else False    
    return Course
```

```python
### resultList = []
rdf = pd.DataFrame(columns=['RegistrationNum', 'CourseID', 'CourseName', 'Semester', 'Credit', 'isMajor', 'isLab', 'GPA', 'LetterGrade'])
for df in dfList:
    row = 1 if df.iloc[1, 0] != None else 2
    for row in range(row, df.shape[0]):
        for column in range(3, df.shape[1]):
            s = df.iloc[0, column]
            if((s != None) and (len(s) >= 6) and (s[3].isnumeric())):
                result = dict()
                result["RegistrationNum"] = df.iloc[row, 0]
                
                course = getCourse(s)
                result["CourseID"] = course['CourseID']
                result["CourseName"] = course['CourseName']
                result["Semester"] = course['Semester']
                result["Credit"] = course['Credit']
                result["isMajor"] = course['isMajor']
                result["isLab"] = course['isLab']
                
                gpaString = df.iloc[row, column]
                if gpaString == '':
                    continue
                result["GPA"] = float(gpaString.split()[0])
                result["LetterGrade"] = gpaString.split()[1]
#                 resultList.append(result)
                try:
                    filt = (rdf['RegistrationNum'] == result["RegistrationNum"]) & (rdf['CourseID'] == result["CourseID"])
                    rdf.drop(index = rdf.loc[filt].index, inplace = True)
                except:
                    co = 0
                finally:
                    rdf = rdf.append(result, ignore_index = True)

# rdf = pd.DataFrame(resultList)
# rdf.set_index("RegistrationNum", inplace = True)
```
